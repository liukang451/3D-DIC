```python
import predic as dm

dic = dm.DIC_NR()
dic.set_parameters("img_gen/ref50.bmp", "img_gen/def50.bmp", 11, [0, 0])
results = dic.calculate()

#print(results)
```

```python
results.shape
```

```python
count =0

for lines in results:
    for elems in lines:
        for dots in elems:            
            if dots != 0:
                count +=1
print("there are",count,"pixles changed")
        
```
