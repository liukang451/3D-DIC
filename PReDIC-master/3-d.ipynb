```python
# a 3d function that runs 2d function twice, 4 images as imput,return two matrix respecticely
import predic as dm

def two_D_Function(aR,aD):
    #img_gen/ref50.bmp img_gen/def50.bmp
    dic = dm.DIC_NR()
    dic.set_parameters(aR, aD, 11, [0, 0])
    results = dic.calculate()
    return results

def three_D_Function(aR,aD,bR,bD):
    re1=two_D_Function(aR,aD)
    re2=two_D_Function(bR,bD)
    return [re1,re2]

three_D_Function("ref50.bmp","ref50.bmp","ref50.bmp","ref50.bmp")

#todo overlap
```
